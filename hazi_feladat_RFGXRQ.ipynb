Készleten lévő mennyiségek adatai:

```sql
SELECT MIN(KESZLET) AS 'Legkisebb készleten lévő mennyiség', 
       MAX(KESZLET) AS 'Legnagyobb készleten lévő mennyiség', 
       ROUND(AVG(KESZLET),0) AS 'Átlagos készleten lévő mennyiség'
FROM Termek
```

Listaár adatok:

```sql
SELECT MIN(LISTAAR) AS 'Legkisebb listaár', 
       MAX(LISTAAR) AS 'Legnagyobb listaár', 
       ROUND(AVG(LISTAAR),0) AS 'Átlagos listaár',
       ROUND(STDEV(LISTAAR),0) AS 'Listaár-eloszlás', 
       ROUND(VAR(LISTAAR),0) AS 'Listaár variancia'
FROM Termek
WHERE LISTAAR IS NOT NULL
AND LEN(LISTAAR) > 0
```

3 legtöbbet használt mérési egység:

```sql
SELECT TOP 3
        MEGYS AS 'Mérési egység',
        COUNT(*) AS 'Termékek száma'
FROM Termek
WHERE MEGYS IS NOT null
GROUP BY MEGYS
ORDER BY COUNT(*) desc
```
